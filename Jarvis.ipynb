```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, StorageContext 
from llama_index.readers.docling import DoclingReader
from llama_index.core import Settings
from llama_index.llms.ollama import ollama
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core.node_parser import MarkdownElementNodeParser

reader = DoclingReader()

loader = SimpleDirectoryReader(input_dir="C:\Deepak\Deepak Mohan\AI\EKPO Extract.xlsx", file_extractor=reader)


llm = ollama(model="llama3.2:latest", request_timeout=60, settings=Settings(temperature=0.1, max_tokens=1000))
embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-MiniLM-L6-v2", trust_remote_code=True)

Settings.embed_model = embed_model

Settings.llm = llm

node_parser = MarkdownElementNodeParser()

index = VectorStoreIndex.from_documents(documents=loader, transformations=[node_parser],show_progress=True)

result = index.query("What is the EKPO Extract?")
#, llm=llm, similarity_top_k=3)

print(result)
```
